```python
#Оценка уровня строительства и дизайна недвижимости от года постройки
plt.figure(figsize=(10, 6))
plt.scatter(df['yr_built'], df['grade'], alpha=0.5)
plt.xlabel('Год постройки')
plt.ylabel('Оценка уровня строительства и дизайна (grade)')
plt.title('Оценка уровня строительства и дизайна недвижимости от года постройки')
plt.grid(True)
plt.show()
```

```python
# Построение круговой диаграммы.  Pаспределение домов от наличия вида на набережную
plt.figure(figsize=(6, 6))
labels = ['Без видовой линии', 'С видовой линией']
sizes = df['waterfront'].value_counts()
colors = ['#abcdef', '#00bb00']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Распределение домов от наличия видовой линии на набережную')
plt.axis('equal')
plt.show()
```

```python
# Построение гистограммы
plt.figure(figsize=(10, 6))
plt.hist(df[df['waterfront'] == 0]['price'], bins=30, alpha=0.5, label='Без видовой линии', color='blue')
plt.hist(df[df['waterfront'] == 1]['price'], bins=60, alpha=1, label='С видовой линией', color='orange')
plt.xlabel('Стоимость (в млн. $)')
plt.ylabel('Количество')
plt.title('Распределение стоимости недвижимости от наличия видовой линии')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='view', hue='waterfront', data=df)
plt.xlabel('Оценка вида')
plt.ylabel('Количество')
plt.title('Связь между видом на набережную и оценкой вида')
plt.legend(title='Видовая линия', labels=['Без', 'С'])
plt.grid(True)
plt.show()
```

Домов с видом на набережную всего 0.8процента, и наличие вида на набережную кратно увеличивает параметр VIEW(как видовой квартиры), особого бонуса в цене таких квартир не наблюдаю.

```python
# Изучите распределение этажей домов.
# Построение столбчатой диаграммы.
plt.figure(figsize=(8, 6))
df['floors'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
plt.title('Распределение этажей домов')
plt.grid(True)
plt.show()
```

```python
#Изучите распределение состояния домов
plt.figure(figsize=(8, 6))
df['condition'].value_counts().sort_index().plot(kind='bar', color='lightgreen')
plt.xlabel('Состояние домов')
plt.ylabel('Количество домов')
plt.title('Распределение состояния домов')
plt.grid(True)
plt.show()
```

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
