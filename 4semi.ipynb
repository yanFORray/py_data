```python
df = pd.read_csv("kc-house-data.csv")
df.head(10) 
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5)
plt.xlabel('Жилая площадь (кв. футы)')
plt.ylabel('Стоимость (в млн. $)')
plt.title('Стоимость недвижимости от жилой площади')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['yr_built'], df['price'], alpha=0.5)
plt.xlabel('Год постройки')
plt.ylabel('Стоимость (в млн. $)')
plt.title('Стоимость недвижимости от года постройки')
plt.grid(True)
plt.show()
```

```python
#Оценка уровня строительства и дизайна недвижимости от года постройки
plt.figure(figsize=(10, 6))
plt.scatter(df['yr_built'], df['grade'], alpha=0.5)
plt.xlabel('Год постройки')
plt.ylabel('Оценка уровня строительства и дизайна (grade)')
plt.title('Оценка уровня строительства и дизайна недвижимости от года постройки')
plt.grid(True)
plt.show()
```

```python
# Построение круговой диаграммы.  Pаспределение домов от наличия вида на набережную
plt.figure(figsize=(6, 6))
labels = ['Без видовой линии', 'С видовой линией']
sizes = df['waterfront'].value_counts()
colors = ['#ff909a', '#acb3f0']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Распределение домов от наличия видовой линии на набережную')
plt.axis('equal')
plt.show()
```
