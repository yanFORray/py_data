1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0
1.3 Создайте признаки года продажи, месяца продажи
1.4 Удалите признаки date, zipcode, lat, long

```python
import pandas as pd
df = pd.read_csv("2sem.csv")
df.head(10) 
```

```python
# 1.2 Создать новый признак delta_renovated
df["delta_renovated"] = df.apply(lambda row: row["yr_renovated"] - row["yr_built"] if row["yr_renovated"] >= row["yr_built"] else 0, axis=1)


# 1.3 Создать признаки года продажи и месяца продажи
df["year_of_sale"] = pd.to_datetime(df["date"], format="%Y%m%dT%H%M%S").dt.year
df["month_of_sale"] = pd.to_datetime(df["date"], format="%Y%m%dT%H%M%S").dt.month

# 1.4 Удалить признаки date, zipcode, lat, long
df.drop(["date", "zipcode", "lat", "long"], axis=1, inplace=True)

# Вывод обновленного DataFrame с новыми признаками
print(df.head(5))
```

```python
# 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

# Используем метод pivot_table для создания сводной таблицы с минимальными, средними и максимальными ценами в зависимости от состояния дома
summary_prices_by_condition = df.pivot_table(values='price', index='condition', aggfunc=['min', 'mean', 'max'])

print("\nМинимальная, средняя и максимальная стоимости домов в зависимости от состояния:")
print(summary_prices_by_condition)
```

```python
# 3.4 Каких домов в зависимости от этажности и количества спален больше?

# Используем метод pivot_table для создания сводной таблицы с подсчетом количества домов в зависимости от этажности и количества спален
count_by_floors_bedrooms = df.pivot_table(values='price', index='floors', columns='bedrooms', aggfunc='count')

print("\nТаблица с подсчетом количества домов в зависимости от этажности и количества спален:")
print(count_by_floors_bedrooms)
```
