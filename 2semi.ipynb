Найдите количество пропущенных ячеек в данных

```python
# Изучение типов данных
data_types = df.dtypes
print("Типы данных:")
print(data_types)

# Подсчет пропущенных значений
missing_values = df.isnull().sum()
print("\nКоличество пропущенных значений:")
print(missing_values)

# Основные статистики
statistics = df.describe()
print("\nОсновные статистики:")
print(statistics)

```

```python
# 3.1 В каком диапазоне изменяются стоимости недвижимости?
min_price = df["price"].min()
max_price = df["price"].max()
print(f"Минимальная стоимость недвижимости: {min_price}")
print(f"Максимальная стоимость недвижимости: {max_price}")

# 3.2 Какую долю в среднем занимают жилая площадь от всей площади по всем домам?
average_residential_ratio = (df["sqft_living"] / df["sqft_lot"]).mean()
print(f"Средняя доля жилой площади от всей площади: {average_residential_ratio:.2f}")

# 3.3 Как много домов с разными этажами в данных?
floors_count = df["floors"].nunique()
print(f"Количество домов с разным количеством этажей: {floors_count}")

# 3.4 Насколько хороши состояния у домов в данных?
conditions_count = df["grade"].value_counts()
print("Количество домов в каждом состоянии:")
print(conditions_count)

# 3.5 Найдите года, когда построили первый дом, когда построили последний дом в данных?
earliest_year = df["yr_built"].min()
latest_year = df["yr_built"].max()
print(f"Год постройки первого дома: {earliest_year}")
print(f"Год постройки последнего дома: {latest_year}")
```

```python
# 4.1 Сколько в среднем стоят дома, у которых 2 спальни?
average_price_2_bedrooms = df[df["bedrooms"] == 2]["price"].mean()
print(f"Средняя стоимость домов с 2 спальнями: {average_price_2_bedrooms:.2f}")

# 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?
average_area_high_price = df[df["price"] > 600000]["sqft_lot"].mean()
print(f"Средняя общая площадь домов с ценой выше 600,000: {average_area_high_price:.2f}")

# 4.3 Как много домов коснулся ремонт?

renovated_houses = df[df["yr_renovated"] > 0]
count_renovated_houses = len(renovated_houses)
print(f"Количество домов с ремонтом: {count_renovated_houses}")

# 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?
average_price_high_grade = df[df["grade"] > 10]["price"].mean()
average_price_low_grade = df[df["grade"] < 4]["price"].mean()
price_difference = average_price_high_grade - average_price_low_grade
print(f"Разница в средней цене между домами с grade > 10 и grade < 4: {price_difference:.2f}")
```

Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?
